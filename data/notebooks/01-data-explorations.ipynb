Keep:
* users with ≥ 1 game
* games with ≥ 5 users

```python
min_users_per_game = 5

df = pd.read_csv("../raw/trictac_data_avis.csv")
df = df[['Game', 'Avatar_Name', 'Score']]

popular_games = (
    df.groupby("Game")["Avatar_Name"]
      .nunique()
      .loc[lambda x: x >= min_users_per_game]
      .index
)

df_clean = df[df["Game"].isin(popular_games)]
```

```python
matrix = df_clean.pivot_table(
    index="Game",
    columns="Avatar_Name",
    values="Score",
    aggfunc="mean"
)
```
